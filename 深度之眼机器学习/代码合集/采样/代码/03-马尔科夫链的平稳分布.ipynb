### 马尔可夫平稳收敛验证

```python
import numpy as np
```

```python
transfer_matrix = np.array([[0.6,0.2,0.2],[0.3,0.4,0.3],[0,0.3,0.7]], dtype=np.float32)
start_matrix = np.array([[0.5, 0.2, 0.3]], dtype=np.float32)
for i in range(30):
    print(start_matrix)
    start_matrix = np.dot(start_matrix, transfer_matrix)
```
